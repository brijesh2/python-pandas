```python
import pandas as pd
df = pd.read_excel("cross.xlsx")
df
```

```python
pd.crosstab(df.nationality,df.handedness)
```

```python
pd.crosstab(df.sex,df.handedness)
```

```python
# when use margins then it shows the total
pd.crosstab(df.nationality,df.handedness, margins=True)
```

```python
 pd.crosstab([df.sex],[df.nationality,df.handedness])
```

```python
 pd.crosstab([df.nationality,df.sex],[df.handedness])
```

```python
pd.crosstab(df.sex,df.handedness,margins=True,margins_name='Total')
```

```python
pd.crosstab(df.sex,df.handedness,margins=True,margins_name='Total',normalize='index')
```

```python
import numpy as np
pd.crosstab([df.sex],[df.handedness],values=df.age , aggfunc=np.sum)
```

```python
import numpy as np
pd.crosstab([df.sex],[df.handedness],values=df.age , aggfunc=np.mean)
```
