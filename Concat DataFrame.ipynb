```python
india_weather = pd.DataFrame({
    "city":['delhi','mumbai','pune','goa'],
    "humidity":[2,4,6,8],
    "temperature":[23,20,12,10]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city":['newyork','delhi','mumbai','orland'],
    "humidity":[3,5,6,8],
    "temperature":[34,21,20,10]
})
us_weather
```

```python
df = pd.merge(india_weather,us_weather,on='city',how='outer',indicator=True)
df
```

```python
df = pd.merge(india_weather,us_weather,on='city',how='outer',indicator=True , suffixes=('_1','_2'))
df
```
